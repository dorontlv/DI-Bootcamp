```python
Exercise 1 : Small Quizz


What is the key difference between traditional machine learning and deep learning?

In traditional machine learning, you typically design the features manually and then train the model.
In deep learning, the neural networks automatically learn features from the raw data, and so you don't need to manually select features.
In deep learning, you guide the model by specifying the type of output (if it's categories or labels), but it has the ability to independently identify and understand the underlying patterns in the data.

# ~~~~~~~~~~~~~~~~~~~~

How do artificial neural networks (ANNs) mimic the human brain?

ANNs mimic the human brain by using layers.  In these layers you have nodes (like neurons) that process information.
The neurons learn patterns and make decisions, by using weights and connections to other neurons.
This is very similar to how biological neurons operate in the brain.

# ~~~~~~~~~~~~~~~~~~~~

Why does deep learning perform better on large datasets compared to traditional machine learning?

Deep learning does better job with large datasets because neural networks can learn complex patterns and features more effectively if the amount of data is bigger.
On the other hand, traditional machine learning will not cope well in such case.

# ~~~~~~~~~~~~~~~~~~~~

What are some challenges of deep learning, and how can they be addressed?

Data requirements: It needs a large amount of data to perform well.
Computational costs: Training the neural network can be expensive because these might be long computations.
Interpretability: Understanding why the model made specific decisions is very hard.

# ~~~~~~~~~~~~~~~~~~~~

What is feature engineering, and why is it not needed in deep learning?

Feature engineering is the process of manually selecting, manipulating and transforming raw data into meaningful inputs for the machine learning models.
It requires our decision to identify which features are relevant.
In deep learning, feature engineering is not needed because neural networks automatically learn and extract features directly from the data.

# ~~~~~~~~~~~~~~~~~~~~

What role do hidden layers play in a deep learning model?

Hidden layers in deep learning are responsible for feature extraction and pattern learning.
Hidden layers and neurons use weights, biases, and activation functions.
Each level is responsible for a more abstract representation of the data.

# ~~~~~~~~~~~~~~~~~~~~

In an artificial neural network (ANN), what is the function of an activation function?

Activation functions introduce non-linearity into the network, and so the model can learn more complex relationships in the data.
Otherwise there would be only linear transformations, and that would limit its abilities.
Activation functions also help control the output range of the data.
```

```python
'''
Exercise 3 : Building a Simple Neural Network with TensorFlow/Keras

Build a simple neural network using TensorFlow/Keras to classify handwritten digits from the MNIST dataset.

The network should have:
One input layer.
One hidden layer with 128 neurons and ReLU activation.
One output layer with 10 neurons (for 10 classes) and softmax activation.

Dataset: The MNIST dataset is included in TensorFlow/Keras and can be loaded using:
from tensorflow.keras.datasets import mnist

'''


from tensorflow import keras
from tensorflow.keras.datasets import mnist

# load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# normalize the input data to the range [0, 1]
x_train = x_train / 255.0
x_test = x_test / 255.0

# build the neural network model
model = keras.models.Sequential([
    keras.layers.Reshape(target_shape=(28*28,), input_shape=(28, 28)),  # Flatten the input
    keras.layers.Dense(128, activation='relu'),  # Hidden layer with 128 neurons
    keras.layers.Dense(10, activation='softmax')  # Output layer with 10 neurons for 10 digit classes
])

# Compile the model
# This is like the example on the Octopus
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',  # Suitable for integer class labels
              metrics=['accuracy'])


# Train the model
model.fit(x_train, y_train, epochs=5, batch_size=32)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Test accuracy: {test_accuracy:.4f}")
```

```python
'''
Exercise 5 : Implementing Forward and Backward Propagation in Python
You will code a simple neural network that performs forward propagation and backpropagation for a regression problem (predicting exam scores based on study hours).

'''


import numpy as np

# Initialize input data (features)
x = np.array([4, 80])  # 4 hours studied, previous test score: 80

# Initialize weights and bias
w = np.array([0.6, 0.3])  # Initial weights
b = 10  # Initial bias

# Forward Propagation
def forward_propagation(x, w, b):
    z = np.dot(x, w) + b  # Weighted sum
    return z  # Linear activation (No ReLU here, it's a regression task)

# Compute prediction
y_pred = forward_propagation(x, w, b)
y_true = 85  # Actual exam score

# Compute Loss (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2

# Compute Gradients
grad_w = -(y_true - y_pred) * x  # Partial derivatives with respect to weights
grad_b = -(y_true - y_pred)  # Partial derivative with respect to bias

# Update Weights and Bias
learning_rate = 0.01
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

# Print Results
print("Initial Prediction:", y_pred)
print("Loss:", loss)
print("Updated Weights:", w_new)
print("Updated Bias:", b_new)


'''

Run the code and observe how the weights and bias update.
Explain why updating weights using gradient descent reduces the error.
Modify the initial weights or learning rate and see how it affects learning.


Answer:
--------

Updating weights using gradient descent reduces the error, because it adjusts the weights in the direction that minimizes the loss.
The loss measures the difference between the predicted values and the true values.
This is working because the gradient is the derivative of the loss function with respect to the weights.

learning_rate controls the rate of change.
If you change the learning_rate, then the change rate of w and b will be different.

'''
```

```python
'''

Exercise 6 : Visualizing Predictions on the MNIST Dataset

Train a simple neural network using TensorFlow/Keras on the MNIST dataset. After training, visualize some of the predictions made by the model.
Dataset: The MNIST dataset is included in TensorFlow/Keras.

Here are the steps for this exercise :

Load the MNIST dataset
Normalize the data
One-hot encode the labels
Build the model
Compile the model
Train the model
Make predictions
Visualize some predictions

'''


from tensorflow.keras.datasets import mnist
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

# load the MNIST dataset (using TensorFlow/Keras)
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# normalize pixel values to the range 0-1
x_train = x_train / 255.0
x_test = x_test / 255.0

# convert labels into one-hot encoded format
y_train = tf.one_hot(y_train, 10)
y_test = tf.one_hot(y_test, 10)

# define the sequential model
model = keras.models.Sequential([
    keras.layers.Reshape(target_shape=(28*28,), input_shape=(28, 28)),  # flatten the input.
    keras.layers.Dense(128, activation='relu'),  # Hidden layer with 128 neurons
    keras.layers.Dense(10, activation='softmax')  # Output layer with 10 neurons for 10 digit classes
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',  # Since labels are one-hot encoded
              metrics=['accuracy'])


# train the model (Fit the model on the training data)
model.fit(x_train, y_train, epochs=5, batch_size=32)

# make predictions
predictions = model.predict(x_test)

# convert predictions to class labels
predicted_labels = np.argmax(predictions, axis=1)

# evaluate the model on the test data
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Test accuracy: {test_accuracy:.4f}")

# let's plot the first 20 images
plt.figure(figsize=(15, 10))  # set the size of the figure
for i in range(20):
    plt.subplot(4, 5, i+1)  # create a grid of 4 rows x 5 columns
    plt.imshow(x_test[i], cmap='gray')  # display the image
    real_label = np.argmax(y_test[i])  # get the true class label from the one-hot encoded format
    predicted_label = predicted_labels[i]  # get the predicted label
    plt.title(f"Real: {real_label}\nPred: {predicted_label}")  # put the labels as title
    plt.axis('off')  # turn off the axis

plt.tight_layout()  # adjust spacing between images
plt.show()

```
